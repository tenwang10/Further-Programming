```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/AI & Data Science BootCamp/sales_returns.csv")
data.head()
```

```python
data.info()
```

```python
X = data.loc[:,["cust_age","cust_gender","sub_category","city"]]
y = data.returned
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
X.cust_gender = enc.fit_transform(X.cust_gender)
X.sub_category = enc.fit_transform(X.sub_category)
X.city = enc.fit_transform(X.city)
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X,y) # training, learning, fitting
```

```python
model.predict([[27,0,13,156]])  # this order has high prob of returning, Block it
```

```python
model.predict([[24,1,10,100]])  # Good, ship it.
```

```python
model.score(X,y)
```
